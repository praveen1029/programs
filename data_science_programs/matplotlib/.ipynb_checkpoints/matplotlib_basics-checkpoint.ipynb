```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(0,10)
y = np.arange(0,20,2)
plt.plot(x,y)
plt.savefig("chart.png",transparent = True)
plt.show()
```

```python
x = np.arange(0,10)
y = np.arange(0,20,2)
plt.plot(x,y,c="r",ls="dashdot")
plt.grid()
plt.show()
```

```python
x1 = np.arange(1,11)
y1 = np.arange(11,21)
x2 = np.arange(1,11)
y2 = np.arange(0,20,2)
plt.plot(x1,y1,x2,y2)
plt.grid()
plt.show()
```

```python
x = np.arange(0,181,2)
y = np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
x = [2,3.5,5,2]
y = [2,3.5,2,2]
plt.plot(x,y)
plt.show()
```
