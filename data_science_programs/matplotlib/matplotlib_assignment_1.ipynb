```python
import pandas as pd
from matplotlib import pyplot as plt 
from matplotlib.pyplot import figure
```

```python
# Exercise 1: Read Total profit of all months and show it using a line plot

df = pd.read_csv("company_sales_data.csv")
x = df["month_number"]
y = df["total_profit"]
plt.plot(x,y)
plt.xlabel("Month Number")
plt.ylabel("Total profit")
plt.show()
```

```python
# Exercise Question 2: Get Total profit of all months and show line plot with the following Style properties

df = pd.read_csv("company_sales_data.csv")
x = df["month_number"]
y = df["total_profit"]
plt.plot(x,y,c="r",ls="dotted",marker="o",lw=3,label="profit per month")
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.legend(loc='lower right')
plt.show()
```

```python
# Exercise Question 3: Read all product sales data and show it using a multiline plot

plt_1 = plt.figure(figsize=(10,7)) #change figure size

df = pd.read_csv("company_sales_data.csv")
x = df["month_number"]
y1 = df["facecream"]
y2 = df["facewash"]
y3 = df["toothpaste"]
y4 = df["bathingsoap"]
y5 = df["shampoo"]
y6 = df["moisturizer"]

plt.plot(x,y1,label="facecream")
plt.plot(x,y2,label="facewash")
plt.plot(x,y3,label="toothpaste")
plt.plot(x,y4,label="bathingsoap")
plt.plot(x,y5,label="shampoo")
plt.plot(x,y6,label="moisturizer")
plt.legend(loc="upper left")
plt.xlabel("Month Number")
plt.ylabel("Product Sold")
plt.show()
```

```python
# Exercise Question 4: Read toothpaste sales data of each month and show it using a scatter plot

df = pd.read_csv("company_sales_data.csv")
x = df["month_number"]
y = df["toothpaste"]
plt.scatter(x,y)
plt.grid(linestyle="-")
plt.xlabel("Month Number")
plt.ylabel("Product Sold")
plt.show()
```

```python
# Exercise Question 5: Read face cream and facewash product sales data and show it using the bar chart

plt_1 = plt.figure(figsize=(10,7))
df = pd.read_csv("company_sales_data.csv")
x = df["month_number"]
y1 = df["facecream"]
y2 = df["facewash"]
plt.bar(x,y1,width=.25)
plt.bar(x+.25,y2,width=.25)
plt.xlabel("Month Number")
plt.ylabel("Product Sold")
plt.show()
```

```python
# Exercise Question 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot

plt_1 = plt.figure(figsize=(10,7))
df = pd.read_csv("company_sales_data.csv")
x = df["month_number"]
y = df["bathingsoap"]
plt.bar(x,y)
plt.xlabel("Month Number")
plt.ylabel("Product Sold")
plt.show()
```

```python
# Exercise Question 7: Read the total profit of each month and show it using the histogram to see most common profit ranges

plt_1 = plt.figure(figsize=(10,7))
df = pd.read_csv("company_sales_data.csv")
x = df["total_profit"]
plt.hist(x)
plt.xlabel("Month Number")
plt.ylabel("Product Sold")
plt.show()
```

```python
# Exercise Question 8: Calculate total sale data for last year for each product and show it using a Pie chart

plt_1 = plt.figure(figsize=(10,7))
df = pd.read_csv("company_sales_data.csv")
pie_label = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
x = [df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(), df['bathingsoap'].sum(), df['shampoo'].sum(), df['moisturizer'].sum()]
plt.pie(x,labels= pie_label,autopct='%1.0f%%')
plt.show()
```

```python
# Exercise Question 9: Read Bathing soap facewash of all months and display it using the Subplot

plt_1 = plt.figure(figsize=(10,7)) #change figure size
df = pd.read_csv("company_sales_data.csv")
x = df["month_number"]
y1 = df["facewash"]
y2 = df["bathingsoap"]
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(x,y1, label = 'Bathingsoap Sales Data')
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(x,y2, label = 'Face Wash Sales Data')
axarr[1].set_title('Sales data of  a facewash')
plt.show()

# f, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
# ax1.plot(x,y1, label = 'Bathingsoap Sales Data')
# ax1.set_title('Sales data of  a Bathingsoap')
# ax2.plot(x,y2, label = 'Face Wash Sales Data')
# ax2.set_title('Sales data of  a facewash')
# plt.show()
```
